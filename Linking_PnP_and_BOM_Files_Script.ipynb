```python
import numpy as np # imports the numpy library
import pandas as pd # imports the pandas library for dataframe generation
import openpyxl # imports the library for xlsx file
import sys # imports the library for system info

print(sys.version)
# Reads the BOM csv file
bomFile = pd.read_csv("bom.csv")
print("BOM File")
print(bomFile)
print("\n")
# Reads the PNP excel file
pnpFile = pd.read_excel("RedBoard-top.xlsx")
# Adds an LCSC column to the PNP file
pnpFile['LCSC'] = 'N/A'
print("PNP File")
print(pnpFile)
print("LCSC Values")
# Initializes arrays of LCSC values and Designator values
numCols_BOM = bomFile.shape[0]
values_LCSC = []
values_Design = []
all_Designators = []
# Stores LCSC values from BOM file into array
for lcscElement in bomFile['LCSC']:
    values_LCSC.append(lcscElement)
#print(values_LCSC)
#print("\n")
# Stores Designator values from BOM file into array
designIndex = 0
for designElement in bomFile['Designator']:
    # Splits arrays with multiple designator elements
    designArray = designElement.split(",")
    # Stores these designator elements individually in a normal array
    for designArrayVal in designArray:
        all_Designators.append(designArrayVal)
    # Stores these values in the corresponding 2D array
    values_Design.insert(designIndex, designArray)  
    # Increments the design index by 1
    designIndex = designIndex + 1
#print(values_Design)
#print("\n")
#print(all_Designators)
#print("\n")
# Initializes a dictionary object for LCSC keys and Designator values
values_LCSC_Design_Dict = {}
lcsc_row = 0
# Traverses the LCSC array and stores Designator values for the corresponding LCSC key
for ind_LCSC in values_LCSC:
    values_LCSC_Design_Dict[ind_LCSC] = values_Design[lcsc_row]
    # Increments the row number by 1
    lcsc_row = lcsc_row + 1
print("Dictionary Object for LCSC keys and Designator values")
print(values_LCSC_Design_Dict)
# Creates a PNP file duplicate for easy accessibility
pnpDuplicate = np.array(pnpFile)
rowStorage = []
# Traverses all designators
for elementDes in all_Designators:
    # Stores the row number index for each designator
    rowNum = np.where(pnpDuplicate == elementDes)[0][0]
    # Adds this index to the row storage array
    rowStorage.append(rowNum)
print(rowStorage)
# Suppresses the SettingWithCopy warning
pd.options.mode.chained_assignment = None
# Initializes designator index
designatorIndex = 0
# Traverses through all keys in the dictionary (Linking Process)
for key in values_LCSC_Design_Dict:
    # Looks at each individual designator in the dictionary
    for designator in values_LCSC_Design_Dict[key]:
        # Provides the LCSC value for each designator
        pnpFile["LCSC"][rowStorage[designatorIndex]] = key
        # Incremenets the designator index by 1
        designatorIndex = designatorIndex + 1
print("New PNP File")
print(pnpFile)
# Traverses through all row numbers in the row storage array to group based on LCSC (Grouping Process)
print("Grouped PNP File")
pnpCopy = np.array(pnpFile)
for rowIndex in rowStorage:
    print(pnpCopy[rowIndex])
# Rearranges the rows based on the increasing order of the LCSC part number (Sorting Process)
# Retrieves the list of keys
keyList = list(values_LCSC_Design_Dict.keys())
# Initializes two empty arrays for key list copies
keyListCopy = []
finalKeyCopy = []
# Traverses through the key list
for key in keyList:
    # Converts the string value without the letter into an int
    keyListCopy.append(int(key[1:]))
    # Sorts the integer values
    keyListCopy.sort()
# Traverses through the key list copy with sorted integers
keyIndex = 0
for key in keyListCopy:
    # Assembles the original key in the sorted format
    newKey = keyList[keyIndex][0:1] + str(key)
    # Adds the new key to the final key list
    finalKeyCopy.append(newKey)
print(finalKeyCopy)
row_LCSC_Design_Dict = {}
newRowStorage = []
pnpMoreDuplicate = np.array(pnpFile)
# Traverses the LCSC keys in the final key array
for ind_LCSC in finalKeyCopy:
    # Finds the row numbers corresponding to the LCSC keys
    location = list(np.where(pnpMoreDuplicate == ind_LCSC)[0])
    # Provides these row numbers as values for each LCSC key
    row_LCSC_Design_Dict[ind_LCSC] = location
# Traverses the LCSC keys in the newly constructed dictionary containing LCSC keys and corresponding row values
print("Sorted PNP File")
for keyVal in row_LCSC_Design_Dict:
    # Looks at each row number in the value array for each key
    for rowNumber in row_LCSC_Design_Dict[keyVal]:
        # Prints the corresponding row number from the PNP file
        print(pnpMoreDuplicate[rowNumber])
```
